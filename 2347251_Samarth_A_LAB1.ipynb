```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("/content/UScereal (1).csv")
```

```python
#Bar plot for 2 variables
plt.figure(figsize=(8, 5))
plt.bar(data['mfr'], data['calories'], color='blue')
plt.xlabel('Manufacturer (mfr)')
plt.ylabel('Calories')
plt.title('Bar Chart with 2 Variables mfr and Calories')
plt.show()
```

```python
#barplot with 3 variables
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
bar1 = ax.bar(data['mfr'], data['calories'], bar_width, label='Calories', color='blue')
bar2 = ax.bar(data['mfr'], data['protein'], bar_width, label='Protein', color='orange', bottom=data['calories'])
ax.set_xlabel('Manufacturer (mfr)')
ax.set_ylabel('Values')
ax.set_title('Bar Chart with 3 Variables mfr, Calories, and Protein')
ax.legend()

plt.show()
```

```python
# Scatter plot with 2 variables
plt.figure(figsize=(8, 5))
plt.scatter(data['sugars'], data['calories'], c='blue', alpha=0.5)
plt.xlabel('Sugars')
plt.ylabel('Calories')
plt.title('Scatter Plot with 2 Variables Sugars and Calories')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(data['sugars'], data['calories'], c=data['protein'], cmap='plasma', s=data['protein']*10, alpha=0.6)
plt.colorbar(scatter, label='Protein')
plt.xlabel('Sugars')
plt.ylabel('Calories')
plt.title('Scatter Plot with 3 Variables Sugars, Calories, and Protein')
plt.grid(True)
plt.show()
```

```python
#2 variables histogram on calories and sodium
plt.hist(data['calories'], bins=25, alpha=0.45, color='red')
plt.hist(data['sodium'], bins=25, alpha=0.45, color='blue')
plt.title("Histogram of Calories and Sodium")
plt.legend(['Calories', 'Sugars'])
plt.show()
```

```python
# Histogram with 3 variables
#2 variables histogram on calories and sodium
plt.hist(data['protein'], bins=25, alpha=0.45, color='red')
plt.hist(data['fibre'], bins=25, alpha=0.45, color='blue')
plt.hist(data['carbo'], bins=25, alpha=0.45, color='green')
plt.title("Histogram of Protien, Fibre and Carbo")
plt.legend(['Protien', 'Fibre', 'Carbo'])
plt.show()
```

```python
#Pie chart to count the maximum occurance/production in UScereal
mfr_counts = data['mfr'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(mfr_counts, labels=mfr_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribution of Manufacturers')
plt.show()
```

```python
#pair plot
selected_variables = ['calories', 'protein', 'fat', 'sugars']
sns.pairplot(data[selected_variables])
plt.suptitle('Pair Plot', y=1.02)
plt.show()
```

```python
# Extracting data
x = data['calories']
y = data['protein']
sizes = data['fat'] * 10
colors = data['fat']
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.title('Calories vs Protein with Fat as Size')
plt.colorbar(label='Fat')
plt.show()
```

```python
#boxplot
plt.figure(figsize=(10,8))
sns.boxplot(x='shelf', y='calories', data = data, palette='tab20b')
plt.title('Box Plot of Calories distribution on Shelf')
plt.xlabel('Shelf')
plt.ylabel('Calories')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(10, 8))
sns.violinplot(x='shelf', y='sodium', data=data, palette='bright')
plt.title('Violin Plot for Sodium across on Shelves')
plt.xlabel('Shelf')
plt.ylabel('Sodium')
plt.show()
```

```python
#Area Plot
variables = ['calories', 'protein', 'fat']
grouped_df = data.groupby('shelf')[variables].mean()
plt.figure(figsize=(10, 6))
plt.stackplot(grouped_df.index, grouped_df.values.T, labels=variables)
plt.xlabel('Shelf')
plt.ylabel('Nutritional Component')
plt.title('Area Plot of Calories, Protien, Fat Components by Shelf')
plt.legend(loc='upper left')
plt.show()
```

```python
# Density
plt.figure(figsize=(10, 6))
sns.kdeplot(data['calories'], shade=True)
plt.title('Calories in Cereal Products')
plt.xlabel('Calories')
plt.show()
```
